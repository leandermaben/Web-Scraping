# Twitter Scraping

```python
import bs4 as bs
import urllib.request
import pandas as pd
from twitterscraper import query_tweets
from twitterscraper import query_tweets_from_user
from twitterscraper.query import query_user_info
import datetime as dt
```

## Selecting Dataset of Users who Tweeted on Maharashtra Election

```python
begin=dt.date(2019,11,29)
end=dt.date(2019,11,30)
tweets=query_tweets('Maharashtra election',begindate=begin,enddate=end,lang='english')
```

The data in the form of a data frame.

```python
df=pd.DataFrame(t.__dict__ for t in tweets)
df.head()
```

Getting the users.

```python
name=df['screen_name'].unique()
print(len(name))
```

## Getting user data

Here we get username,fullname,likes,number of followers and following users. 

```python
userData={}
allUsers=[]
for user in name:
    user_info = query_user_info(user)
    if user_info:
        userData["username"]=user_info.user
        userData["fullname"]=user_info.full_name
        userData["likes"]=user_info.likes
        userData["followers"]=user_info.followers
        userData["following"]=user_info.following
        allUsers.append(userData.copy())
```

```python
cols=['username','fullname','likes','followers','following']
data_frame=pd.DataFrame(allUsers,columns=cols)
data_frame.head()
```

## Getting the Most Like Tweets

This is determined based on the number of likes per tweet.

```python
posts=[]
for user in list(data_frame["username"]):
    user_tweets=query_tweets_from_user(user)
    df2=pd.DataFrame(t.__dict__ for t in user_tweets)
    df2.sort_values(by="likes", ascending=False, inplace=True, kind='quicksort', na_position='last')
    posts.append(list(df2["text"][:6]))
```

```python
data_frame["Top Posts"]=posts
data_frame.head()
```
